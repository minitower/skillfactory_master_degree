## Задание 1

Даны 3 вектора, необходимо исследовать их на линейную зависимость:
$$\vec{a_1} = (1; 8; -1), \vec{a_2}=(-2;3;3), \vec{a_3}=(4;-11;9)$$

В первую очередь запишем данные вектора в виде матрицы:
$$\begin{pmatrix}
1 & -2 & 4 \\ 
8 & 3 & -11 \\
-1 & 3 & 9  
\end{pmatrix}$$

Посчитаем определитель этой матрицы. Если определитель не равен 0, то векторы линейно независимы:
$$det=\begin{vmatrix}
1 & -2 & 4 \\ 
8 & 3 & -11 \\
-1 & 3 & 9  
\end{vmatrix} = (1*3*9)+(-2*-11*-1)+(4*8*3)-(4*3*-1)-(1*-11*3)-(-2*8*9) = 290$$
Т. к. определитель матрицы не равен 0, следовательно, ранг матрицы равен 3, следовательно, вектора линейно независимы

В первую очередь исследуем то, что мы можем исследовать: линейную зависимость имеющихся вектором (без учета последней координаты). Тогда, найдем определитель матрицы:

Определитель этой матрицы равен 0 ~~(повезло)~~, что значит, что вектора уже линейно зависимые.

## Задание 3

Дана матрица. Нужно выразить ее определитель:
$$\begin{pmatrix}
1 & -1 & a & -1 \\ 
-1 & -2 & b & -1 \\
-2 & 0 & c & 1  \\
0 & -1 & d & 0
\end{pmatrix}$$

Т. к. перед нами матрица 4x4 использовать какую-то готовую формулу будет проблематично. Поэтому используем правильно разложения определителя по строке. В соответствии с этим правилом определитель будет равен:
$$det = 1*M_{11}-(-1)*M_{12}+a*M_{13}-(-1)*M_{14}$$

Вычислим миноры:
$$M_{11} = \begin{vmatrix}
-2 & b & -1 \\
0 & c & 1  \\
-1 & d & 0
\end{vmatrix} = (-2*c*0)+(b*1*-1)+(-1*0*d)-(-1*c*-1)-(-2*1*d)-(b*0*0) = 0-b+0-c+2d-0 = -b-c-2d$$

$$M_{12} = \begin{vmatrix}
-1 & b & -1 \\
-2 & c & 1  \\
0 & d & 0
\end{vmatrix} = (-1*c*0)+(b*1*0)+(-1*-2*d)-(-1*c*0)-(-1*1*d)-(b*-2*0) = 0+0+2d-0+d-0 = 3d$$

$$M_{13} = \begin{vmatrix}
-1 & -2 & -1 \\
-2 & 0 & 1  \\
0 & -1 & 0
\end{vmatrix} = (-1*0*0)+(-2*1*0)+(-1*-2*-1)-(-1*0*0)-(-1*1*-1)-(-2*-2*0) = 0+0-2+0-1-0=-3$$

$$M_{14} = \begin{vmatrix}
-1 & -2 & b \\
-2 & 0 & c  \\
0 & -1 & d
\end{vmatrix} = (-1*0*d)+(-2*c*0)+(b*-2*-1)-(b*0*0)-(-1*c*-1)-(-2*-2*d) = 0+0+2b-0-c-4d = 2b-c-4d$$

Теперь подставим найденные миноры в формулу для определителя:
$$det = -b-c+2d+3d-3a+2b-c-4d$$

После упрощения:
$$det = b-2c+d-3a$$

**Ответ:**

Определитель равен $det = b-2c+d-3a$

# Задание 7

Написанная функция показывает количество пропусков в каждом столбце. Согласно ней, пропуски отсутствуют, следовательно, пропущенных значений нет.

```python
import pandas as pd
import numpy as np
import plotly.graph_objects as go

df = pd.read_csv('./data/data.xls')

for i in df.columns:
    print(f'Количество пропущенных значений в столбце {i}',df[i].loc[df[i].isna()==True].count())
```

# Задание 8

```python
data_matrix = df.drop('cnt', axis=1).values
dependent = []

if len(data_matrix[0]) > np.linalg.matrix_rank(data_matrix):
    print('Есть зависимые столбцы!')
    for i in range(len(data_matrix[0])):
        data_matrix_del = np.delete(data_matrix, i, axis=1)
        if len(data_matrix_del[0]) == np.linalg.matrix_rank(data_matrix_del):
            dependent.append(i)
print('Линейно зависимые столбцы: ', dependent)

data_matrix = np.delete(data_matrix, dependent[1], axis=1)

print(f'Удален столбец {dependent[1]}')
if len(data_matrix[1]) == np.linalg.matrix_rank(data_matrix):
    print('Теперь столбцы линейно независимы')
    
df = df.drop(df.columns[dependent[1]], axis=1)
```

Мы нашли 3 столбца, которые выражаются друг через друга: 9, 10, 11. Было принято решение удалить столбец 9. После этого ранг матрицы стал равен количеству столбцов, что говорит, что столбцы линейно независимы

# Задание 10

```python
X = df.drop('cnt', axis=1).values
X = np.column_stack((np.ones(len(X)), X))
y = df['cnt'].values

w_hat = np.linalg.inv(X.T@X)@X.T@y
print('Коэффициенты линейной регрессии:', w_hat)

y_pred = X@w_hat
print('MAE: ', abs(y-y_pred).sum()/len(y))
print('MAPE: ', sum((abs((y-y_pred)/y)))*100/len(y))
```

Была построена модель линейной регрессии с помощью numpy. Оценка качества модели линейной регрессии происходила при помощи 2 методов: средней абсолютной ошибки прогноза (MAE), а также среднего процента ошибки (MAPE). В результате стало видно, что модель ошибается в среднем на 45.5%, что является очень большим показателем. Следовательно, модель нуждается в доработке.
