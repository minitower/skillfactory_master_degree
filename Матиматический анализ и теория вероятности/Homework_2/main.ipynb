# Задача 1 (ручное решение)

1) В условии задачи сказано, что присутствуют **два вида систем**: с подключением к "Умному дому" и без таковой. Т. к. других вариантов не дано, то будем считать, что выбрав любой товар мы можем отнести его к одной из категорий, при этом только к одной (иначе говоря, что множества непересекаются и являются полной группой событий). Если предположение верно, то можно выразить вероятности:
$$p_{ум} = 1/4=0.25$$
$$p_{без ум} = 3/4=0.75$$
2) Далее необходимо выбрать распределение, согласно которому будет происходить моделирование. Т. к. выборки независимы друг от друга (конечно, это возможно только в выборке с возвращением, либо в случае, когда количество товаров равно бесконечности, однако из-за того, что информации о количестве товаров нет будем считать, что они независимы), есть лишь 2 возможных исхода (с подключением к умному дому и без него), то можно говорить о том, что распределение **биноминальное**. Биноминальное распределение может быть выражено как:
$$P(X=k)=С_n^m p^m q^{n-m}$$
3) Рассчитаем вероятности, что не более 2 экземпляров будут доступны для подключения. Следовательно, у нас есть 3 возможных исхода:
- Не будет ни одного товара, который бы имел подключение к умному дому;
- Имеется 1 товар, который имеет подключение к умному дому;
- Имеется 2 товара, которые имеют подключение к умному дому.
Учитывая предпосылки пункта 1 о том, что множества не пересекаются мы можем применить формулу суммы вероятностей для поиска необходимой вероятности. Таким образом, необходимо вычислить вероятности для вышеперечисленных функций, после чего их сложить:
$$P(X \leq 2)=P(X=0)+P(X=1)+P(X=2)$$
Подставим необходимые значения в формулу из пункта 2:
$$P(X=0)=C_0^7(0.25)^0(0.75)^7=(0.75)^7=0.1335$$
$$P(X=1)=C_1^7(0.25)^1(0.75)^6=7*0.25(0.75)^6=0.3115$$
$$P(X=2)=C_2^7(0.25)^2(0.75)^5=21*(0.25)^2(0.75)^5=0.3118$$
4) Сложив вероятности, получим ответ:
$$P(X \leq 2)=0.1335+0.3115+0.3118=0.7568$$
**Ответ:**
Вероятность того, что среди 7 экземпляров будет не более двух с подключением к умному дому составляет 75.68%

```python
# Решение на Python
import scipy.stats as stats

p_with = 0.25
p_without = 0.75

a = stats.binom.cdf(2, 7, 0.25)
print ('Вероятность того, что среди 7 экземпляров будет не более двух с подключением к умному дому составляет ', a)
```

Найдите пропущенную вероятность для X = 0, математическое ожидание и дисперсию. Решите задачу "вручную" и с помощью python.

# Задача 2 (решение вручую)
1) Первое, что необходимо для решения данной задачи - это внести некоторые предположения о данном распределении. На мой взгляд, их 2:
- Распределение является полной группой событий (без этого невозможно посчитать необходимые данные). Т. е. не может быть больше 5 публикаций;
- События являются независимыми. Т. е. не может быть случая, когда публикаций и 1, и 2 одновременно в одной группе.
С учетом этих предположений можно говорить о том, что $P(X=0)$ будет равна:
$$\sum_{i=0}^nP_i=P(X=0)+P(X=1)+P(X=2)+P(X=3)+P(X=4)+P(X=5)=1$$
Тогда $P(X=0)$ найти не сложно:
$$P(X=0)=1-P(X=1)+P(X=2)+P(X=3)+P(X=4)+P(X=5)=1-0.3+0.3+0.2+0.1+0.1=1-1=0$$
Следовательно, вероятность того, что количество публикаций равно 0 в данной группе каналов равна 0
2) Рассчитаем мат. ожидание по формуле:
$$E(X)=\sum_i^n X_i P(X_i)$$
Подставим значения в формулу:
$$E(X)=(0*0)+(1*0.3)+(2*0.3)+(3*0.2)+(4*0.1)+(5*0.1)=2.4$$
3) Рассчитаем дисперсию по формуле:
$$D(X)=E(X^2)-E(X)^2$$
Рассчитаем $E(X^2)$:
$$E(X^2)=(0^2*0)+(1^2*0.3)+(2^2*0.3)+(3^2*0.2)+(4^2*0.1)+(5^2*0.1)=7.4$$
Подставим значения в формулу:
$$D(X)=7.4-2.4^2=7.4-5.76=1.64$$
**Ответ**
Таким образом:
1) Вероятность отсутствия публикаций равно 0;
2) Мат. ожидание случайной величины равно 2.4;
3) Дисперсия случайной величины равна 1.64

```python
# Решение на Python
import numpy as np

X = np.array([0, 1, 2, 3, 4, 5])
P = np.array([0.3, 0.3, 0.2, 0.1, 0.1])

P = np.insert(P, 0, 1-sum(P))
print(P)

expectation = np.sum(X * P)

expectation_sqr = np.sum((X**2) * P)

var = expectation_sqr - expectation**2

print(f"P(X=0): {P[0]}")
print(f"Математическое ожидание: {expectation}")
print(f"Дисперсия: {round(var, 2)}")
```

# Задача 3 (решение вручную)
## Вопрос 1
1) В первую очередь стоит задаться вопросом о том, какое распределение использовать в данном случае. С учетом предпосылки о независимости, а также о 2 возможных исходах (ноутбук сломан, ноутбук не сломан), а также малая вероятность и большое количество выборки, то можно говорить о **распределении Пуассона**. При этом $\lambda = np$. Следовательно, $\lambda = 10000*0.0002 = 2$
2) Распределение Пуассона выглядит следующим образом:
$$P(X=k)=\frac{\lambda^k*e^{- \lambda}}{k!}$$
Следовательно, для нахождения вероятности единственного сломанного ноутбука $k$ будет равно 1:
$$P(X=1)=\frac{2^1*e^{-2}}{1!}=2*0.1353=0.2706$$
## Вопрос 2 пункт а
1) Распределение остается тем же. С учетом независимости (как было сказано в условии) рассчитаем вероятность поломки 5 и менее ноутбуков, после чего вычтем из полной вероятности. Таким образом:
$$P(X=0)=\frac{2^0*e^{-2}}{0!}=1*0.1353=0.1353$$
$$P(X=1)=\frac{2^1*e^{-2}}{1!}=2*0.1353=0.2706$$
$$P(X=2)=\frac{2^2*e^{-2}}{2!}=2*0.1353=0.2706$$
$$P(X=3)=\frac{2^3*e^{-2}}{3!}=1.333*0.1353=0.1804$$
$$P(X=4)=\frac{2^4*e^{-2}}{4!}=0.667*0.1353=0.0902$$
$$P(X=5)=\frac{2^5*e^{-2}}{5!}=0.267*0.1353=0.0361$$
$$P(X \leq 5) = 0.1353+0.2706+0.2706+0.1804+0.0902+0.0361 = 0.9832$$
$$P(X > 5) = 1 -  0.9832 = 0.0168$$
Следовательно, вероятность того, что за месяц сломается более 5 ноутбуков равна 0.0168

## Вопрос 2 пункт б
1) Для того, чтобы сделать это необходимо рассчитать математическое ожидание для распределения Пуассона, которое равно: 
$$E(X)=\lambda$$
Таким образом ожидаемое количество поломок равно 2.
Далее расчитаем математическое ожидание расходов как: 
$$2*500=1000$$
Следовательно, ожидаемые расходы составляют 1000 долларов в месяц

```python
# Решение на Python
n=10_000
p=0.0002

question_1 = stats.poisson.pmf(1, n*p)
print('Вероятность сломать лишь один ноутбук равна ', round(question_1, 4))

question_2a = 1-stats.poisson.cdf(5, n*p)
print('Вероятность того, что сломается больше 5 ноутбуков ', round(question_2a, 4))

cost = 500
question_2b = stats.poisson.stats(n*p, moments='m')*cost
print('Мат. ожидание расходов на ремонт составляет: ', question_2b)
```
