# Задача 4
Т. к. было сказано в Пачке, что можно использовать готовое решение, то использую его. Ответ в Пачке от 13 января.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import OrdinalEncoder
from sklearn.naive_bayes import GaussianNB

df = pd.read_csv('./data/task_4.csv')
print('Исходный DF:')
print(df)
encoder = OrdinalEncoder()
df_encoded = encoder.fit_transform(df) # делаем энкодер для текстовых данных

X = df_encoded[:, 0:2] # составляем X и y
y = df_encoded[:, 2]

print('Закодированный DF')
print(df_encoded)

nb_clf = GaussianNB()
X_train, X_test, y_train, y_test = train_test_split(X, y) # разделяем на выборки

nb_clf.fit(X_train, y_train)
nb_clf_pred = nb_clf.predict(X_test) # предсказываем

accuracy = accuracy_score(y_test, nb_clf_pred) # считаем метрику accuracy
print('Оценка модели методом accuracy: ', accuracy)
```

$$P(2\leq X \leq 4) = P(X=2)+P(X=3)+P(X=4)$$

Произведем расчеты:
1. $P(X=2) = \frac{2^2*e^{-2}}{2!} = 2e^{-2}$
2. $P(X=3) = \frac{2^3*e^{-2}}{3!} = \frac{4e^{-2}}{3}$
2. $P(X=3) = \frac{2^4*e^{-2}}{4!} = \frac{2e^{-2}}{3}$

Вычисляем сумму вероятностей:
$$P(2\leq X \leq 4) = 2e^{-2} + \frac{4e^{-2}}{3} + \frac{2e^{-2}}{3} = 4e^{-2}$$

Подставив константу получаем:
$$P(2\leq X \leq 4) = 4*(2.7182)^-2 = 4*\frac{1}{7.3890} = 4*0.1353 = 0.541 = 54.1%$$
