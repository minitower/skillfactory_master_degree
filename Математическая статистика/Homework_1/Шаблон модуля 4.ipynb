```python
'''Ваш ответ на задание №2 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''
mean = 3.005  
true_mean = 3.00 
std = 0.015
n = 50
alpha = 0.05

z_statistic = (mean - true_mean) / (std / math.sqrt(n))

z_critical = stats.norm.ppf(1 - alpha / 2)
reject_null = abs(z_statistic) > z_critical
p_value = 2 * (1 - stats.norm.cdf(abs(z_statistic)))

print('Расчетная z-статистика: ', z_statistic)
print('Критичная z-статистика: ', z_critical)
if abs(z_statistic) > z_critical:
    print('Сравнение z-статистики с ее критичным уровнем говорит о том, что нулевая гипотеза отвергается')
elif abs(z_statistic)<=z_critical:
    print('Сравнение z-статистики с ее критичным уровнем говорит о том, что нулевая гипотеза принимается')
print('p-value: ', p_value)
print('alpha: ', alpha)
if p_value<alpha:
    print('Сравнение p-value с уровнем значимости говорит о том, что нулевая гипотеза отвергается')
elif p_value>=alpha:
    print('Сравнение p-value с уровнем значимости говорит о том, что нулевая гипотеза принимается')
```

**Выводы:**

Используем z-статистику для данного распределения, т. к. у нас известно отклонение, которое задал производитель. Далее, исходя из расчитаной z-статистики можно говорить о том, что нулевая гипотеза *отвергается* и дозатор совершает ошибки. Об этом также говорит p_value, который меньше уровня значимости (который составляет *0.05*).

Задание 4 (про токи)

```python
'''Ваш ответ на задание №4 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

old_arr = np.array([80.1, 82.3, 84.1, 82.6, 85.3, 81.3, 83.2, 81.7, 82.2, 81.4])
new_arr = np.array([80.7, 81.3, 84.6, 81.7, 86.3, 84.3, 83.7, 84.7, 82.8, 84.4, 85.2, 84.9])

n1 = len(old_arr)
n2 = len(new_arr)

mean_old = np.mean(old_arr)
mean_new = np.mean(new_arr)

std_old = np.std(old_arr, ddof=1)
std_new = np.std(new_arr, ddof=1)

alpha=0.05
df = n1 + n2 - 2

var = ((n1 - 1) * std_old**2 + (n2 - 1) * std_new**2) / (n1 + n2 - 2)

SE = np.sqrt(var * (1 / n1 + 1 / n2))

t_statistic = (mean_old - mean_new) / SE
t_critical = stats.t.ppf(1 - alpha/2, df=df)

p_value = 2 * stats.t.sf(abs(t_statistic), df=df)

print('t-критерий рассчитанный: ', t_statistic)
print('t-критерий критический: ', t_critical)
if t_statistic>t_critical:
    print('Так как t-критерий больше критического значения, нулевая гипотеза отвергается')
elif t_statistic<=t_critical:
    print('Так как t-критерий меньше критического значения, нулевая гипотеза принимается')
print('p-value: ', p_value)
print('alpha: ', alpha)
if p_value<alpha:
    print('Так как p-value меньше alpha, то нулевая гипотеза отвергается')
elif p_value>=alpha:
    print('Так как p-value больше alpha, то нулевая гипотеза принимается')
```

# Выводы
1) Использовался t-критерий, т. к. надо было найти разницу в двух выборках;
2) В результате анализа было выявлено, что нулевая гипотеза принимается, а, следовательно, разницы между средними математическими ожиданиями данных величин нет. Поэтому можно заменить старую конструкцию схемы на новую, т. к. она пропускает тот же средний ток.

# Выводы
1) Использовали парный t-критерий ввиду того, что исследовались 2 зависимые выборки;
2) В ходе исследования видно, что нулевая гипотеза отвергается, следовательно существуют статистически значимые различия между двумя выборками, что говорит о том, что эффект от приема препаратов присутствует.

Задание 8 (про фрукты и IQ)

```python
'''Ваш ответ на задание №8 
- верно указана применяемая статистика и обоснование ее применения
- получено расчетное значение статистики
- получено критическое значение статистики
- получено значение p_value
- сделан статистический вывод на основе сравнения критического значения с расчетным
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''

orange_group = [125, 138, 112, 130, 124, 125, 110, 112, 117, 126]
pear_group = [124, 102, 142, 130, 131, 124, 127, 115, 108, 109]
alpha = 0.05

var_group_orange = np.var(orange_group, ddof=1)
var_group_pear = np.var(pear_group, ddof=1)

if var_group_orange>var_group_pear:
    F_statistic = var_group_orange / var_group_pear 
else:
    F_statistic = var_group_pear / var_group_orange

df_group_1 = len(orange_group) - 1
df_group_2 = len(pear_group) - 1

alpha = 0.05
F_critical = stats.f.ppf(1 - alpha / 2, df_group_1, df_group_2)
p_value = 2 * stats.f.sf(abs(F_statistic), df_group_1, df_group_2)

print('F-критерий рассчитанный: ', F_statistic)
print('F-критерий критический: ', F_critical)
if F_statistic>F_critical:
    print('Так как F-критерий больше критического значения, нулевая гипотеза отвергается')
elif F_statistic<=F_critical:
    print('Так как F-критерий меньше критического значения, нулевая гипотеза принимается')
print('p-value: ', p_value)
print('alpha: ', alpha)
if p_value<alpha:
    print('Так как p-value меньше alpha, то нулевая гипотеза отвергается')
elif p_value>=alpha:
    print('Так как p-value больше alpha, то нулевая гипотеза принимается')
```

```python
'''Ваш ответ на задание №12
- верно указана применяемая статистика и обоснование ее применения
- расчет выполнен через последовательные суммы квадратов
- получены расчетные значения статистики
- получены значения p_value
- сделан статистический вывод на основе сравнения p-value и уровня значимости
- сделан содержательный вывод по поставленной проблеме'''
```

```python
data = [[440, 438, 429, 502, 602], [308, 328, 326, 326, 325],
        [892, 868, 855, 866, 932], [737, 798, 876, 810, 861],
        [1575, 849, 759, 1602, 1327], [1001, 901, 958, 1032, 883]]
alpha = 0.05

data_groups = [np.array(values) for values in data]
group_len = [len(values) for values in data]

total_mean = np.mean([value for group in data_groups for value in group])
group_means = [np.mean(values) for values in data_groups]

SST = sum((value - total_mean) ** 2 for group in data_groups for value in group)
SSR = sum(n * np.sum((values - np.mean(values)) ** 2)
                for n, values in zip(group_len, data_groups))
SSE= sum(n * (mean - total_mean) ** 2 for n, mean in zip(group_len, group_means))

df_between = len(data_groups) - 1
df_within = sum(group_len) - len(data_groups)

F_statistic = (SSE / df_between) / (SSR / df_within)

F_critical = stats.f.ppf(1 - alpha, df_between, df_within)
p_value = stats.f.sf(F_statistic, df_between, df_within)

print('F-критерий рассчитанный: ', F_statistic)
print('F-критерий критический: ', F_critical)
if F_statistic>F_critical:
    print('Так как F-критерий больше критического значения, нулевая гипотеза отвергается')
elif F_statistic<=F_critical:
    print('Так как F-критерий меньше критического значения, нулевая гипотеза принимается')
print('p-value: ', p_value)
print('alpha: ', alpha)
if p_value<alpha:
    print('Так как p-value меньше alpha, то нулевая гипотеза отвергается')
elif p_value>=alpha:
    print('Так как p-value больше alpha, то нулевая гипотеза принимается')
```
